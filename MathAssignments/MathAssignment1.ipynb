# Tehtävä 3

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as la
```

```python
# Luodaan x ja y matriisit, sekä X matriisi x:n avulla.
x = np.arange(1, 11)
y = np.array([14, 12, 34, 24, 43, 38, 41, 57, 55, 50])

X = np.vstack([np.ones(len(x)), x]).T

print(X)
```

```python
# Ratkaistaan regressiokertoimet
M = la.inv(X.T @ X) @ X.T @ y
print(f"Regressiokertoimet: {M[0]:.3} ja {M[1]:.3}")
```

```python
xx = np.linspace(1, 10, 1000)
XX = np.vstack([np.ones(len(xx)), xx]).T
yy = XX.dot(M)
plt.plot(x, y, 'o')
plt.plot(xx, yy)
```

# Tehtävä 4

```python
# Luodaan matriisi X = [1, x, x^2]
X = np.vstack([np.ones(len(x)), x, x**2]).T
print(X)
```

```python
# Ratkaistaan regressiokertoimet
M = la.inv(X.T @ X) @ X.T @ y
print(f"Regressiokertoimet: {M[0]:.6}, {M[1]:.6} ja {M[2]:.5}")
```

```python
# Piirretään kuvaaja
xx = np.linspace(1, 10, 1000)
XX = np.vstack([np.ones(len(xx)), xx, xx**2]).T
yy = XX.dot(M)
plt.plot(x, y, 'o')
plt.plot(xx, yy)
```
